## 3. Prepare Dataset

```python
transactions = transactions[['t_dat', 'customer_id', 'article_id']]
```

```python
transactions.head(1)
```

```python
transactions['combi'] = transactions['t_dat'].astype(str) + "," + transactions['customer_id']
transactions.head(1)
```

```python
combi_list = new_trans['combi'].tolist()
```

```python
%%time 

# Obtain all transactions of customers who purchased popular item
transactions = transactions[transactions['combi'].isin(combi_list)]

transactions.head(1)
```

```python
%%time 
trans_dict = {}

for date, item, cust in zip(transactions['t_dat'], transactions['article_id'], transactions['customer_id']):
    
    # Check if transaction date is present
    if date in trans_dict:
        today_trans = trans_dict[date]
        
        if cust in today_trans:
            today_trans[cust].append(str(item))
            
        else:
            today_trans[cust] = [str(item)]
        
    else:
        trans_dict[date] = {cust: [str(item)]}
```

```python
%%time
final_trans = []
for key in trans_dict:
    curr_date = trans_dict[key]
    for cust in curr_date:
        if len(curr_date[cust]) < 2:
            continue
        else:
            final_trans.append(curr_date[cust])
```

```python
len(final_trans)
```

## 4. Run Apriori

Possibly Change Min_Support to 0.0002

```python
from apyori import apriori
association_rules = apriori(final_trans, min_support=0.004, min_confidence=0.1, min_lift=1, use_colnames=True)
association_results = list(association_rules)
```

```python
for i in range(0, len(association_results)):
    print(association_results[i][0])
```

```python
len(association_results)
```

```python
item_basket = []
support = []
confidence = []
lift = []
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    rule = tuple(items)
    item_basket.append(rule)
    support.append(item[1])
    confidence.append(item[2][0][2])
    lift.append(item[2][0][3])
```

```python
apriori_df = pd.DataFrame(list(zip(item_basket, support, confidence, lift)),
                         columns = ['Items', 'Support', 'Confidence', 'Lift'])
```

```python
apriori_df
```

## 5. Check if Items Make Sense

```python
association_df = apriori_df.copy()
```

```python
prod_list = []
section_list = []
garment_list = []

for items in association_df['Items']:
    if len(items) == 2:
        curr_prod = (articles[(articles['article_id'] == int(items[0])) | (articles['article_id'] == int(items[1]))]
              ['prod_name'].values)    
        curr_sec = (articles[(articles['article_id'] == int(items[0])) | (articles['article_id'] == int(items[1]))]
              ['section_name'].values)
        curr_gar = (articles[(articles['article_id'] == int(items[0])) | (articles['article_id'] == int(items[1]))]
              ['garment_group_name'].values)
    elif len(items) == 3:
        curr_prod = (articles[(articles['article_id'] == int(items[0])) | (articles['article_id'] == int(items[1]))
                             | (articles['article_id'] == int(items[2]))]
              ['prod_name'].values)  
        curr_sec = (articles[(articles['article_id'] == int(items[0])) | (articles['article_id'] == int(items[1]))
                             | (articles['article_id'] == int(items[2]))]
              ['section_name'].values)
        curr_gar = (articles[(articles['article_id'] == int(items[0])) | (articles['article_id'] == int(items[1]))
                            | (articles['article_id'] == int(items[2]))]
              ['garment_group_name'].values)
    else:
        curr_prod = (articles[(articles['article_id'] == int(items[0]))]
              ['prod_name'].values)    
        curr_sec = (articles[(articles['article_id'] == int(items[0]))]
              ['section_name'].values)
        curr_gar = (articles[(articles['article_id'] == int(items[0]))]
              ['garment_group_name'].values)
    
    prod_list.append(list(curr_prod))
    section_list.append(list(curr_sec))
    garment_list.append(list(curr_gar))
```

```python
association_df['Product Names'] = prod_list
association_df['Section Names'] = section_list
association_df['Garment Group'] = garment_list
```

```python
import gc 
```

```python
df = transactions[['customer_id','article_id']]
df.customer_id = [int(x, 16) for x in df.customer_id.str[-16:].values]
df.article_id = df.article_id.astype('int32')
_ = gc.collect()
```

```python
vc = df.article_id.value_counts()
pairs = {}
for j,i in enumerate(vc.index.values):
    USERS = df.loc[df.article_id==i.item(),'customer_id'].unique()
    vc2 = df.loc[(df.customer_id.isin(USERS))&(df.article_id!=i.item()),'article_id'].value_counts()
    
    if len(vc2) == 2:
         pairs[i.item()] = [vc2.index[0], vc2.index[1]]
    elif len(vc2) < 2:
        continue
    else:
        pairs[i.item()] = [vc2.index[0], vc2.index[1], vc2.index[2]]
```
